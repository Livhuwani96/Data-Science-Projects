```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

‚óè Import insurance.csv into your notebook (Source).

```python
# Load the dataset
insurance_data = pd.read_csv('insurance.csv')

# Display the first few rows of the dataframe
insurance_data.head()
```

Plot a scatter plot with age on the x-axis and charges on the y-axis.

```python
# Scatter plot with age on the x-axis and charges on the y-axis

plt.figure(figsize=(10, 6))
plt.scatter(insurance_data['age'], insurance_data['charges'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter Plot of Age vs Charges')
plt.show()
```

Using linear_model.LinearRegression() from sklearn, fit a model to
your data, and make predictions on the data.

```python
# Prepare the data for linear regression
X = insurance_data[['age']]  # Feature
y = insurance_data['charges']  # Target

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
predictions = model.predict(X)
```

Plot another scatter plot with the best-fit line.

```python
# Plot another scatter plot with the best-fit line

plt.figure(figsize=(10, 6))
plt.scatter(insurance_data['age'], insurance_data['charges'], alpha=0.5, label='Data')
plt.plot(insurance_data['age'], predictions, color='red', label='Best-fit line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter Plot of Age vs Charges with Best-fit Line')
plt.legend()
plt.show()
```
