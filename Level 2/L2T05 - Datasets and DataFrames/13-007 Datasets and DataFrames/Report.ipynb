```python
# Import pandas

import pandas as pd
```

```python
# Load data from balance.txt into a DataFrame
df = pd.read_csv('balance.txt', delim_whitespace=True)

```

```python
# Compare the average income based on ethnicity
avg_income_by_ethnicity = df.groupby('Ethnicity')['Income'].mean()

# On average, do married or single people have a higher balance?
avg_balance_by_marital_status = df.groupby('Married')['Balance'].mean()

# What is the highest income in our dataset?
highest_income = df['Income'].max()

# What is the lowest income in our dataset?
lowest_income = df['Income'].min()

# How many cards do we have recorded in our dataset?
total_cards = df['Cards'].sum()

# How many females do we have information for vs how many males?
gender_counts = df['Gender'].value_counts()
```

```python
# Display results
print("Average Income by Ethnicity:")
print(avg_income_by_ethnicity)
print("\nAverage Balance by Marital Status:")
print(avg_balance_by_marital_status)
print("\nHighest Income in the Dataset:")
print(highest_income)
print("\nLowest Income in the Dataset:")
print(lowest_income)
print("\nTotal Number of Cards:")
print(total_cards)
print("\nGender Counts:")
print(gender_counts)
```
