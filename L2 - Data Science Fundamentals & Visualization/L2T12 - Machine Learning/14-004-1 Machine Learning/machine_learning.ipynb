1. For each of the following examples describe at least one possible input and 
output. Justify your answers:  
* 1.1 A self-driving car
* 1.2 Netflix recommendation system
* 1.3 Signature recognition
* 1.4 Medical diagnosis


1. Answer here

*Self-driving car:
Input: Images and sensor data from cameras and LIDAR sensors.

Output: Steering angles, acceleration, and braking commands.

Justification: The self-driving car uses the input data to understand its environment, including the position of other vehicles, pedestrians, traffic signs, and the road’s layout. The output controls the car’s actions to navigate safely.

*Netflix recommendation system:
Input: User’s viewing history, ratings, search queries, and time spent on shows.

Output: A list of recommended movies and TV shows.

Justification: The recommendation system analyzes the input to discern patterns in the user’s preferences and suggests content that matches these preferences, aiming to enhance user engagement.

*Signature recognition:
Input: An image of a handwritten signature.

Output: A decision on whether the signature is genuine or forged.

Justification: The system compares the input signature with a database of known signatures using feature extraction and pattern recognition to verify the signer’s identity.

*Medical diagnosis:
Input: Patient symptoms, medical history, lab test results, and imaging data.

Output: A list of possible diagnoses and recommended treatments.

Justification: The diagnosis system uses the input data to identify patterns and correlations with known medical conditions, helping healthcare professionals to make informed decisions about patient care.


2. For each of the following case studies, determine whether it is appropriate to utilise regression or classification machine learning algorithms. Justify your answers:
* 2.1 Classifying emails as promotion or social based on their content and metadata. 
* 2.2 Forecasting the stock price of a company based on historical data and market trends.
* 2.3 Sorting images of animals into different species based on their visual features.
* 2.4 Predicting the likelihood of a patient having a particular disease based on medical history and diagnostic test results. 

2. Answer here
* 2.1. Classifying emails as promotional or social:

Algorithm: Classification
Justification: The task involves categorizing emails into discrete classes (promotional or social), which is a typical classification problem.
* 2.2. Forecasting the stock price of a company:

Algorithm: Regression
Justification: Stock price prediction requires estimating a continuous value (the price) based on historical data, making it a regression task.
* 2.3. Sorting images of animals into different species:

Algorithm: Classification
Justification: This task involves assigning images to predefined categories (species), which is a classic example of a classification problem.
* 2.4. Predicting the likelihood of a patient having a particular disease:

Algorithm: Classification or Regression
Justification: If the outcome is binary (disease or no disease), it’s a classification problem. However, if the task is to predict a probability score (likelihood), it could be approached as a regression problem that outputs continuous values between 0 and 1.


3. For each of the following real-world problems, determine whether it is appropriate to utilise a supervised or unsupervised machine learning algorithm. Justify your answers:
* 3.1 Detecting anomalies in a manufacturing process using sensor data without prior knowledge of specific anomaly patterns. 
* 3.2 Predicting customer lifetime value based on historical transaction data and customer demographics. 
* 3.3 Segmenting customer demographics based on their purchase history, browsing behaviour, and preferences. 
* 3.4 Analysing social media posts to categorise them into different themes. 


3. Answer here
* 3.1.Approach: Unsupervised Learning
Justification: Since there’s no prior knowledge of specific anomaly patterns, unsupervised learning is suitable for identifying unusual patterns or outliers in the data without labeled examples.

* 3.2. Approach: Supervised Learning
Justification: This task involves using historical transaction data and customer demographics (labeled data) to predict a continuous value (lifetime value), which is a regression problem within supervised learning.

* 3.3. Approach: Unsupervised Learning
Justification: The goal is to discover inherent groupings in the data (purchase history, browsing behaviour, and preferences) without predefined labels, making it a clustering problem, which is an unsupervised learning task.

* 3.4. Approach: Supervised or Unsupervised Learning
Justification: If there are predefined themes and labeled examples of posts, supervised learning can be used. However, if themes are not predefined, unsupervised learning algorithms like clustering can be used to discover themes in the data.
