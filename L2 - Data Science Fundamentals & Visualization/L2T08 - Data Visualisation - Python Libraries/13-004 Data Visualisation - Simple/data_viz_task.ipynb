```python
# Import libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load data

df = pd.read_csv('Cars93.csv')
```

### 1. Boxplot

```python
# Boxplot for revs per mile

filtered_data = df[df['Manufacturer'].isin(['Audi', 'Hyundai', 'Suzuki', 'Toyota'])]

# Create the box plot
plt.figure(figsize=(10, 6))
plt.boxplot([filtered_data[filtered_data['Manufacturer'] == manu]['Rev.per.mile'] for manu in ['Audi', 'Hyundai', 'Suzuki', 'Toyota']])
plt.xticks([1, 2, 3, 4], ['Audi', 'Hyundai', 'Suzuki', 'Toyota'])
plt.title('Box Plot of Revs per Mile for Selected Manufacturers')
plt.ylabel('Rev.per.mile')
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**

Based on the box plot image above, the manufacturer with the car that has the highest revs per mile is Toyota. This is indicated by an outlier point that is distinctly higher than all other data points for the manufacturers shown in the plot.


### 2. Histogram

```python
# Histogram of MPG

plt.figure(figsize=(10, 6))
plt.hist(df['MPG.city'], bins=15, alpha=0.5, label='City MPG')
plt.hist(df['MPG.highway'], bins=15, alpha=0.5, label='Highway MPG')
plt.xlabel('Miles Per Gallon (MPG)')
plt.ylabel('Frequency')
plt.title('Histogram of MPG in the City and on the Highway')
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**

Based on the histogram above, it is generally more fuel efficient to drive on the highway than in the city. The histogram shows that the miles per gallon (MPG) values for highway driving (represented in orange) are higher than those for city driving (represented in blue), indicating better fuel efficiency on the highway. This is typically due to the constant speeds and smoother flow of traffic on highways, which allows for more efficient fuel consumption compared to the frequent stops and starts encountered in city driving.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)

plt.figure(figsize=(10, 6))
plt.plot(df['Wheelbase'], df['Turn.circle'], marker='o')
plt.xlabel('Wheelbase (inches)')
plt.ylabel('Turning circle (feet)')
plt.title('Relationship between Wheelbase and Turning Circle')
plt.grid(True)
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
The scatter plot indicates a positive correlation between a car’s wheelbase and its turning circle. As the wheelbase increases, so does the turning circle, meaning larger vehicles need more space to turn. This relationship is visually represented by an upward trend in the plot

### 4. Barplot

```python
# Barplot for mean horsepower

mean_horsepower_by_type = df.groupby('Type')['Horsepower'].mean()

# Create the bar plot
plt.figure(figsize=(10, 6))
mean_horsepower_by_type.plot(kind='bar')
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.title('Mean Horsepower by Car Type')
plt.xticks(rotation=45)
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**

It indicates that larger car types like Large, midsize and Van generally have more horsepower than smaller types like ‘Compact’ and ‘Small’. However, the ‘Sporty’ category has the highest mean horsepower, suggesting that while larger cars tend to be more powerful, other factors like design,type of engine and intended use also play a significant role in a car’s horsepower
