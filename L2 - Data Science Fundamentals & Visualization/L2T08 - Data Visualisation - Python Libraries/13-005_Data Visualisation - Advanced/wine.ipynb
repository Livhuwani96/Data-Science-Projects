# Data Visualisation Task 2

```python
# Import Libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data

wine_data = pd.read_csv('wine.csv')
```

1. Filter the dataset

```python
# Filter the dataset to only contain 'Cabernet Sauvignon', 'Pinot Noir', and 'Chardonnay' wines.

filtered_wine_data = wine_data[wine_data['variety'].isin(['Cabernet Sauvignon', 'Pinot Noir', 'Chardonnay'])]
```

2.  Creat a multi-plot grid

```python
# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Plot histograms for each variety
for i, variety in enumerate(filtered_wine_data['variety'].unique()):
    plt.subplot(1, 3, i+1)
    subset = filtered_wine_data[filtered_wine_data['variety'] == variety]
    sns.histplot(subset['points'], kde=False, bins=20)
    plt.title(f'Histogram of Points for {variety}')
    plt.xlabel('Points')
    plt.ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```
